# Linear Regression

Linear regression is a supervised machine learning algorithm and statistical technique used to predict a continuous, numerical outcome (dependent variable) by modeling the linear relationship with one or more input features (independent variables). It fits a straight line—represented by the formula — that minimizes the sum of squared differences between observed and predicted data points.

### y=mx+c
- y → output (dependent variable)
- x → input (independent variable)
- m → slope (how much y changes when x changes)
- c → intercept (starting value)



### Key Concepts of Linear Regression
- Types: Simple Linear Regression uses one predictor (e.g., house size to predict price), while Multiple Linear Regression uses two or more.
- Best-Fit Line: The algorithm finds the optimal slope (m) and intercept (c) to minimize prediction error.
- Methods: The most common method is Ordinary Least Squares (OLS), which minimizes the sum of squared residuals.
- Goal: To establish a formula that predicts the outcome for new input data.
- Evaluation: The performance is often measured using metrics like R-squared, which shows how well the independent variables explain the dependent variable.

### Common Assumptions
- Linearity: The relationship between features and targets is linear.
- Independence: Observations are independent of each other.
- Homoscedasticity: The residuals (errors) have constant variance.
- Normality: The residuals are normally distributed.

### Applications
- Forecasting: Predicting sales, stock prices, or temperatures.
- Optimization: Determining how different factors, such as advertising spend, affect sales.
- Data Analysis: Understanding the strength of relationships between variables.
